3. K-Means-- Simple handwritten Implementation (Core Logic)

```python
def kmeans_minus_minus(X, k):
    n_samples = X.shape[0]
    # All the initial points are centered
    centers = X.copy()
    
    # Gradually delete until only k centers remain
    while len(centers) > k:
        # Calculate the distance between each pair and find the one that is closest
        dist_mat = np.linalg.norm(centers[:, None] - centers, axis=2)
        # Excluding its own distance of 0
        np.fill_diagonal(dist_mat, np.inf)
        # Find the nearest point and delete it
        idx_to_del = np.argmin(np.min(dist_mat, axis=1))
        centers = np.delete(centers, idx_to_del, axis=0)
    
    # Find the nearest point and delete it
    km = KMeans(n_clusters=k, init=centers, n_init=1)
    return km.fit_predict(X), centers
```

K—means— —, How is this function called?

Let's start with the key points:
The official sklearn does not have a built-in KMeans function. It cannot be directly imported from sklearn. You can only implement it by hand and encapsulate it into a directly callable function, which can be used just like calling a library function.
Here's the complete KMeans that can be directly copied and run - encapsulated functions + call examples + comparison with sklearn KMeans.

I. Complete KMeans- Encapsulated Code (Directly callable)

```python
import numpy as np
from sklearn.cluster import KMeans

class KMeansMinusMinus:
    def __init__(self, n_clusters):
        self.k = n_clusters
        self.labels_ = None
        self.cluster_centers_ = None

    def fit_predict(self, X):
        n_samples = X.shape[0]
        # 1. Complete KMeans- Encapsulated Code (Directly callable)
        centers = X.copy()
        
        # 2. Keep deleting the nearest center points until there are only k left
        while len(centers) > self.k:
            # Calculate the pairwise distance matrix
            dist_mat = np.linalg.norm(centers[:, np.newaxis] - centers, axis=2)
            # Set the self-distance to infinity
            np.fill_diagonal(dist_mat, np.inf)
            # Find the point closest to others and delete it
            min_dist_each = np.min(dist_mat, axis=1)
            del_idx = np.argmin(min_dist_each)
            centers = np.delete(centers, del_idx, axis=0)
        
        # 3. Find the point closest to others and delete it
        km = KMeans(n_clusters=self.k, init=centers, n_init=1)
        self.labels_ = km.fit_predict(X)
        self.cluster_centers_ = km.cluster_centers_
        return self.labels_
```

II. How to Invoke (Exactly the Same usage as sklearn)

```python
# Create test data
X = np.random.rand(200, 2)

# Call KMeans- just like calling sklearn
kmm = KMeansMinusMinus(n_clusters=3)
y_pred = kmm.fit_predict(X)

# Call KMeans- just like calling sklearn
print("tag clustering：", y_pred)
print("center of clustering：\n", kmm.cluster_centers_)
```

IV. Key Explanations
1. There are no ready-made library functions, so you have to encapsulate them yourself. Just copy the code above and it can be used.
2. Fully replicate the syntax of sklearn: __init__ set the number of clusters, fit_predict training + prediction;
3. Principle: First, take all the centers → Iterate and delete the most recent redundant centers → When only k centers remain, then iterate through standard KMeans.
4. It is suitable for small samples, but it will be slow for large data volumes (a full-distance matrix needs to be calculated).
